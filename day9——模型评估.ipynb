```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#假设 y_true 是真实的标签（0是健康，1是患病）
#假设 y_pred 是模型预测的标签
y_true = [0, 0, 1, 0, 1, 1, 0, 1, 0, 0]
y_pred = [0, 0, 1, 0, 0, 1, 0, 1, 0, 1]

# 1. 打印完整的评估报告（Precision， Recall， F1， Support）
#在工程中， 这是最快了解模型全面表现的方法
print("Classification Report:\n")
print(classification_report(y_true, y_pred, target_names=['Healthy','Cancer']))

# 2. 生成混淆矩阵
cm = confusion_matrix(y_true, y_pred)

# 3. 可视化混淆矩阵
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
           xticklabels=['Pred Healthy', 'Pred Cancer'],
            yticklabels=['True Healthy', 'True Cancer'])
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix in Medical Diagnosis')
plt.show()
```
