```python
import tensorflow as tf
from tensorflow.keras import layers,models
```

```python
# 1.加载数据
mnist = tf.keras.datasets.mnist
(x_train, y_train),(x_test, y_test) = mnist.load_data()

# 2.归一化（让数据处于0-1之间）
x_train, x_test = x_train /255.0, x_test / 255.0

# 3.搭建模型结构
model = models.Sequential([
    #将28*28的二维图片“压扁”成 784 长度的一维向量
    layers.Flatten(input_shape=(28,28)),
    
    # 第一个隐藏层： 128个神经元，使用RuLU激活函数 （防止梯度消失）
    layers.Dense(128, activation='relu'),
    
    #Dropout层：随机让20%的神经元下班，防止过拟合
    layers.Dropout(0.2),
    
    #输出层：10个神经元，对应数字0-9
    #使用Softmax 将输出转换为概率分布
    layers.Dense(10, activation='softmax')
])
```

```python
# 配置优化器、损失函数和评价标准
model.compile(optimizer='adam', # 这是一个会自动调节步长的“聪明”梯度下降
              loss='sparse_categorical_crossentropy', # 多分类问题的标配损失函数
              metrics=['accuracy']) # 我们关心的核心指标：准确率

# 开始训练！epochs=5 表示把所有数据看 5 遍
model.fit(x_train, y_train, epochs=5)

# 最后在测试集上“闭卷考试”
model.evaluate(x_test,  y_test, verbose=2)
```

```python
# 1. 评估模型：看看在没见过的数据上表现如何
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)
print(f'\n测试集准确率: {test_acc:.2%}')

# 2. 实际预测：扔一张图给它
predictions = model.predict(x_test)

# 看看第一张图预测结果中，哪个数字概率最高
import numpy as np
predicted_number = np.argmax(predictions[0])    #np.argmax是numpy里用来找最大值所在位置的函数
true_number = y_test[0]

print(f'模型说是: {predicted_number}, 真实答案是: {true_number}')
```
